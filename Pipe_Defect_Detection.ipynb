# Importing the Libraries

```
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Defining function for Pipe defect detection simulation

```
def MonteCarloPipe(num_simu):
    Bool = np.zeros(num_simu)
    for i in range(num_simu):
      instrument_a = [0,1]
      test = np.random.choice(instrument_a,p=[0.4,0.6])
      Bool[i] = (test==0)
      if test ==0:
        instrument_b = [0,1]
        test = np.random.choice(instrument_b,p=[0.3,0.7])
        Bool[i] = (test==0)
        if test == 0:
          instrument_c = [0,1]
          test = np.random.choice(instrument_c,p=[0.2,0.8])
          Bool[i] = (test==0)
    probability = 1-np.mean(Bool)
    return probability
```

# Inputs for Pipe defect detection problem

```
Number_of_simulation = np.arange(1,2000)
probability_defect_detection = np.array([MonteCarloPipe(n) for n in Number_of_simulation])
```

# Ploting the results

```
plt.plot(Number_of_simulation, probability_defect_detection)
plt.xlabel('Number of simulation')
plt.ylabel('probability of defect detection by any instrument')
plt.title('Pipe defect')
plt.legend(['defect detected'])
plt.grid(True)
plt.show()
```
